```python
from json import loads

import numpy as np
import pandas as pd
csv_path = "../output/repositories.csv"
df = pd.read_csv(csv_path)
# df.head()
```

```python
l_rel = []
l_pri = []
l_pul = []
l_all = []
l_clo = []
for rel, pri, pul, all, clo in zip(
        df["releases"],
        df["primaryLanguage"],
        df["pullRequests"],
        df["all_issues"],
        df["closed_issues"]
):
    print(f"rel={rel}, pri={pri}, pul={pul}, all={all}, clo={clo}")
    j_rel = loads(rel.replace("'", '"'))
    try:
        j_pri = loads(pri.replace("'", '"'))
    except AttributeError as e:
        j_pri = {"name": None}
    j_pul = loads(pul.replace("'", '"'))
    j_all = loads(all.replace("'", '"'))
    j_clo = loads(clo.replace("'", '"'))
    l_rel.append(j_rel["totalCount"])
    l_pri.append(j_pri["name"])
    l_pul.append(j_pul["totalCount"])
    l_all.append(j_all["totalCount"])
    l_clo.append(j_clo["totalCount"])

df["releases"] = l_rel
df["primaryLanguage"] = l_pri
df["pullRequests"] = l_pul
df["all_issues"] = l_all
df["closed_issues"] = l_clo
```
